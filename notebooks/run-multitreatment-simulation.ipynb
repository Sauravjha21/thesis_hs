```python
import os.path

import numpy as np
from tqdm import tqdm

from alias import *
from utils import *
```

```python
dfs = []
dgp_factory_class_list = [
    dgp.LinearFactory,
    dgp.QuickBlockFactory,
    dgp.LinearDriftFactory,
    dgp.LinearSeasonFactory,
    dgp.QuadraticFactory,
    dgp.CubicFactory,
    dgp.SinusoidalFactory,
]
num_treatments = np.arange(31) + 1

NUM_ITERS = 250

for num_trt in num_treatments:
    print(f"Number of treatments: {num_trt}", flush=True)
    num_groups = num_trt + 1
    plan = make_multitreat_plan(
        [
            (
                "Complete",
                bal.Complete,
                est.MultiDiffInMeans,
                {"q": [1 / num_groups] * num_groups},
            ),
            (
                "Simple",
                bal.Simple,
                est.MultiDiffInMeans,
                {"q": [1 / num_groups] * num_groups},
            ),
            # ('Simple + OLS', bal.Simple, est.MultiOLS, {'q': [1 / num_groups] * num_groups}),
            (
                "Smith",
                bal.Smith,
                est.MultiDiffInMeans,
                {"q": [1 / num_groups] * num_groups},
            ),
            (
                "BWD(1)",
                bal.MultiBWD,
                est.MultiDiffInMeans,
                {"q": [1 / num_groups] * num_groups},
            ),
            (
                "BWD(0.5)",
                bal.MultiBWD,
                est.MultiDiffInMeans,
                {"q": [1 / num_groups] * num_groups, "phi": 0.5},
            ),
            # ('BWD + OLS', bal.MultiTBD, est.MultiOLS, {'q': [1 / num_groups] * num_groups}),
            (
                "DM",
                bal.MultiDM,
                est.MultiDiffInMeans,
                {"q": [1 / num_groups] * num_groups},
            ),
            # ('DM + OLS', bal.MultiDM, est.MultiOLS, {'q': [1 / num_groups] * num_groups}),
        ]
    )
    dgp_factory_list = [
        factory(N=1000, K=num_trt) for factory in dgp_factory_class_list
    ]
    for dgp_factory in dgp_factory_list:
        dgp_name = type(dgp_factory.create_dgp()).__name__
        print(f"\nDGP name: {dgp_name}", flush=True)
        for it in tqdm(range(NUM_ITERS)):
            filename = f"multitreat-results/{dgp_name}_k{num_trt}_i{it}.csv.gz"
            if not os.path.isfile(filename):
                result = plan.execute(dgp_factory, seed=it * 1001)
                result["iteration"] = it
                result["sample_size"] = num_trt
                result["dgp"] = dgp_name
                result.to_csv(filename, index=False)
                # upload_file(filename, BUCKET)
                dfs.append(result)
```
